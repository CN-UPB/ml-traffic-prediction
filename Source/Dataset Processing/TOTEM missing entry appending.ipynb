```python
import numpy as np
import pandas as pan
import xml.etree.ElementTree as ET

import matplotlib.pyplot as plt

import os
from datetime import timedelta, date, datetime
```

```python
name = r'..\Datasets\GEANTCombined\all_in_one_complete.csv'
df = pan.read_csv(name)
```

```python
import time

start_date = datetime(2005, 1, 1, 0, 30, 0)
current_date = datetime(2005, 1, 1, 0, 30, 0)
end_date = datetime(2005, 4, 29, 16, 45, 0)

dataframes = []
dataframe = pan.DataFrame(columns=['timestamp', 'source', 'destination', 'bandwidth'])
frame_counter = 0

for source in range(1, 24):
    df_source = df[(df.source == source)]
    #print(df_source)
    for destination in range(1, 24):
        df_destination = df_source[(df_source.destination == destination)]
        #print(df_destination)
        print(str(source) + "." + str(destination))
        while current_date <= end_date:
            single_row = df_destination[(df_destination.timestamp == str(current_date))]
            
            if single_row.empty:
                #print(current_date)
                d = {'timestamp': [current_date], 'source': [source], 'destination': [destination], 'bandwidth': [0]}
                
                dataframes.append(pan.DataFrame(data = d))  
                
            current_date = current_date + timedelta(seconds=900)
            
        df_temp = pan.concat(dataframes)
        dataframe = dataframe.append(df_temp)
        dataframes = []
        current_date = datetime(2005, 1, 1, 0, 30, 0)
dataframe.append(df)
```
