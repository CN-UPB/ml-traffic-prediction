```python
import pandas
import numpy
import pmdarima as pm
import matplotlib.pyplot as plt
from TimeSeriesModelCreator_Parallel_talos import TimeSeriesModelCreator_Parallel_talos
```

```python
label_size = 20
```

# Read in dataset and prepare training and testing sets

```python
dataframe = pandas.read_csv(r'..\Datasets\GEANTCombined\all_in_one_complete_appended.csv')
subsets_training = []
subsets_testing = []
for x in range(1,6):
    subsets_training.append(dataframe[(dataframe.source == x) & (dataframe.destination == 11)][['bandwidth']][-(1000+200):-200])
    subsets_testing.append(dataframe[(dataframe.source == x) & (dataframe.destination == 11)][['bandwidth']][-200:])
```

# Creating LSTM and training

```python
creator = TimeSeriesModelCreator_Parallel_talos(2, r'..\Datasets\GEANTCombined\all_in_one_complete_appended.csv')
modelMatch = {}
for x in range(1,6):
    creator.add_new_model(name = 'test'+str(x), nodes = 1, layer = 1, loss='mean_squared_error', optimizer='adam')
    modelMatch[str(x)+'_11'] = 'test'+str(x)
creator.train_model(1, 5, 11, 11, 1000, 200, modelMatch, epoch = 1000, batch_size = 128, shift = 0)
```

# LSTM predictions

```python
LSTMpredictions = []
for x in range(1,6):
    prediction = creator.predict('test'+str(x), subsets_testing[x-1], 0)
    LSTMpredictions.append(prediction)
```

```python
for x in range(0,5):
    fig, ax1 = plt.subplots()
    ax1.tick_params(labelsize=label_size)
    fig.set_size_inches(18, 10)
    plt.plot(subsets_testing[x].values, label='Real values')
    plt.hlines(subsets_testing[x]['bandwidth'].mean(),xmin=0,xmax=200, label='Real mean')
    plt.plot(range(1,200), LSTMpredictions[x], color = 'green', label='LSTM Predictions')
    plt.hlines(LSTMpredictions[x].mean(),xmin=0,xmax=200, color='red', label='Prediction mean')
    print('Mean percentage error: {0}%'.format((100/subsets_testing[x]['bandwidth'].mean())*LSTMpredictions[x].mean()))
    plt.xlabel('15-Minute Steps', fontsize=label_size)
    plt.ylabel('Bandwidth (kbps)', fontsize=label_size)
    ax1.legend(fontsize=label_size)
    plt.savefig('..\..\Documents\Thesis\Pictures\Practical_Examples\LSTMprediction_{0}.pdf'.format(x))
    plt.savefig('..\..\Documents\Thesis\Pictures\Practical_Examples\LSTMprediction_{0}.svg'.format(x))
    plt.show()
```

# Creating and fitting ARIMA

```python
ARIMAmodels = []
ARIMApredictions = []
ARIMAconfInt = []
for x in range(0,5):
    # Fit a simple auto_arima model
    ARIMAmodel = pm.auto_arima(subsets_training[x].values.reshape(1,1000)[0], error_action='ignore', trace=1, seasonal=True, m=4)
    print(ARIMAmodel.get_params())
    # Create predictions for the future, evaluate on test
    preds = []
    conf_ints = []
    for step in range(0,200):
        pred, conf_int = ARIMAmodel.predict(n_periods=1, return_conf_int=True)
        preds.append(pred)
        conf_ints.append(conf_int)
        ARIMAmodel.update(subsets_testing[x][step:step+1])
    ARIMAmodels.append(ARIMAmodel)
    ARIMApredictions.append(preds)
    ARIMAconfInt.append(conf_ints)
```

```python
for x in range(0,5):
    fig, ax1 = plt.subplots()
    ax1.tick_params(labelsize=label_size)
    fig.set_size_inches(18, 10)
    plt.plot(subsets_testing[x].values, label='Real values')
    plt.hlines(subsets_testing[x]['bandwidth'].mean(),xmin=0,xmax=200, label='Real mean')
    plt.plot(range(-1,199), ARIMApredictions[x], color = 'red', label = 'ARIMA Predictions')
    plt.hlines(LSTMpredictions[x].mean(),xmin=0,xmax=200, color='green', label='Prediction mean')
    plt.xlabel('15-Minute Steps', fontsize=label_size)
    plt.ylabel('Bandwidth (kbps)', fontsize=label_size)
    ax1.legend(fontsize=label_size)
    plt.savefig('..\..\Documents\Thesis\Pictures\Practical_Examples\ARIMAprediction_{0}.pdf'.format(x))
    plt.savefig('..\..\Documents\Thesis\Pictures\Practical_Examples\ARIMAprediction_{0}.svg'.format(x))
    plt.show()
```

```python
for x in range(0,5):
    fig, ax1 = plt.subplots()
    ax1.tick_params(labelsize=label_size)
    fig.set_size_inches(18, 10)
    plt.plot(subsets_testing[x].values, label='Real values')
    plt.plot(range(-1,199), ARIMApredictions[x], color = 'red', label='ARIMA Predictions')
    plt.plot(range(1,200),LSTMpredictions[x], color = 'green', label='LSTM Predictions')
    plt.xlabel('15-Minute Steps', fontsize=label_size)
    plt.ylabel('Bandwidth (kbps)', fontsize=label_size)
    ax1.legend(fontsize=label_size)
    plt.savefig('..\..\Documents\Thesis\Pictures\Practical_Examples\AvLprediction_{0}.pdf'.format(x))
    plt.savefig('..\..\Documents\Thesis\Pictures\Practical_Examples\AvLprediction_{0}.svg'.format(x))
    plt.show()
```

```python
OverallARIMAError = []
OverallLSTMError = []
for x in range(0,5):
    LSTMError = []
    ARIMAError = []
    for step in range(0,198):
        LSTMError.append(abs(LSTMpredictions[x][step][0] - subsets_testing[x].values[step+1][0]))
        ARIMAError.append(abs(ARIMApredictions[x][step+1][0] - subsets_testing[x].values[step][0]))
    OverallARIMAError.append(ARIMAError)
    OverallLSTMError.append(LSTMError)
    fig, ax1 = plt.subplots()
    ax1.tick_params(labelsize=label_size)
    fig.set_size_inches(18, 10)
    plt.plot(ARIMAError, color = 'red', label='ARIMA error')
    plt.plot(range(1,199), LSTMError, color = 'green', label='LSTM error')
    plt.xlabel('15-Minute Steps', fontsize=label_size)
    plt.ylabel('Bandwidth (kbps)', fontsize=label_size)
    ax1.legend(fontsize=label_size)
    plt.savefig('..\..\Documents\Thesis\Pictures\Practical_Examples\AvLERRORprediction_{0}.pdf'.format(x))
    plt.savefig('..\..\Documents\Thesis\Pictures\Practical_Examples\AvLERRORprediction_{0}.svg'.format(x))
    plt.show()
```

```python
for x in range(0,5):
    positivChange = True
    LSTMCounter = 0
    ARIMACounter = 0
    for step in range(1,198):
        start = subsets_testing[x].values[step][0]
        change = start - subsets_testing[x].values[step+1][0]
        if(change <= 0):
            positivChange = True
        else:
            positivChange = False
        LSTMChange = start - LSTMpredictions[x][step][0]
        if(positivChange):
            if(LSTMChange <= 0):
                LSTMCounter += 1
        else:
            if(LSTMChange > 0):
                LSTMCounter += 1
        
        start = subsets_testing[x].values[step-1][0]
        change = start - subsets_testing[x].values[step][0]
        if(change <= 0):
            positivChange = True
        else:
            positivChange = False
        ARIMAChange = start - ARIMApredictions[x][step+1][0]
        if(positivChange):
            if(ARIMAChange <= 0):
                ARIMACounter += 1
        else:
            if(ARIMAChange > 0):
                ARIMACounter += 1
    LSTMratio = LSTMCounter / 199
    ARIMAratio = ARIMACounter / 199
    print(str(LSTMratio))
    print(str(ARIMAratio))
    print("-----------")
```

```python
for x in range(0,5):
    fig, ax1 = plt.subplots()
    ax1.tick_params(labelsize=22)
    fig.set_size_inches(10, 10)
    plt.boxplot([OverallARIMAError[x],OverallLSTMError[x]]) 
    ax1.set_xticklabels(['ARIMA', 'LSTM'])
    plt.ylabel('Bandwidth (kbps)', fontsize=22)
    ax1.legend(fontsize=22)
    plt.savefig('..\..\Documents\Thesis\Pictures\Practical_Examples\BoxERRORprediction_{0}.pdf'.format(x))
    plt.savefig('..\..\Documents\Thesis\Pictures\Practical_Examples\BoxERRORprediction_{0}.svg'.format(x))
    plt.show()
```
