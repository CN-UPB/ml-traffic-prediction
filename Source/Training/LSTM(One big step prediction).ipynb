```python
import pandas
import numpy
import pmdarima as pm
import matplotlib.pyplot as plt
from TimeSeriesModelCreator_Parallel_talos import TimeSeriesModelCreator_Parallel_talos
```

# Read in dataset and prepare training and testing sets

```python
dataframe = pandas.read_csv(r'..\Datasets\GEANTCombined\all_in_one_complete_appended.csv')
subsets_training = []
subsets_testing = []
for x in range(1,6):
    subsets_training.append(dataframe[(dataframe.source == x) & (dataframe.destination == 11)][['bandwidth']][-(1000+200):-200])
    subsets_testing.append(dataframe[(dataframe.source == x) & (dataframe.destination == 11)][['bandwidth']][-200:])
```

# Creating LSTM and training

```python
future = 16
creator = TimeSeriesModelCreator_Parallel_talos(1, r'..\Datasets\GEANTCombined\all_in_one_complete_appended.csv')
modelMatch = {}
for x in range(1,6):
    creator.add_new_model(name = 'test'+str(x), nodes = 1, layer = 1, loss='mean_squared_error', optimizer='adam')
    modelMatch[str(x)+'_11'] = 'test'+str(x)
creator.train_model(1, 5, 11, 11, 1000, 200, modelMatch, epoch = 1000, batch_size = 128, shift = future)
```

# LSTM predictions

```python
LSTMpredictions = []
for x in range(1,6):
    prediction = creator.predict('test'+str(x), subsets_testing[x-1], future)
    LSTMpredictions.append(prediction)
```

```python
for x in range(0,5):
    fig, ax1 = plt.subplots()
    fig.set_size_inches(20, 11.4)
    ax1.tick_params(labelsize = 25)
    #plt.plot(subsets_training[x].values)
    plt.plot(subsets_testing[x].values, label='Real values')
    plt.hlines(subsets_testing[x]['bandwidth'].mean(),xmin=0,xmax=200, label='Real mean')
    plt.plot(range(future-1,200+future-1),LSTMpredictions[x], color = 'red', label='Prediction')
    plt.hlines(LSTMpredictions[x].mean(),xmin=0,xmax=200, label='Prediction mean', color='green')
    plt.xlabel('15-Minute Steps', fontsize=20)
    plt.ylabel('Bandwidth (kbps)', fontsize=20)
    plt.legend(fontsize=20)
    plt.savefig('..\..\Documents\Thesis\Pictures\BigStep\LSTM_Big_Step_{0}{1}.pdf'.format(future,x))
    plt.savefig('..\..\Documents\Thesis\Pictures\BigStep\LSTM_Big_Step_{0}{1}.svg'.format(future,x))
    plt.show()
```
